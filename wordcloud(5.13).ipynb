```python
import wordcloud
import matplotlib.pyplot as plt
%matplotlib inline
keywords = {'Hello':35, 'Python':1, 'Crawl':5, 'Visual':3, 'Creat':4}

wc = wordcloud.WordCloud()
cloud = wc.generate_from_frequencies(keywords)

plt.figure()
plt.imshow(cloud)
plt.show()
```

```python
# 한글로 출력시 한글 깨짐을 방지하기 위해서는 font 변경 필요
import wordcloud
import matplotlib.pyplot as plt
%matplotlib inline
keywords = {'안녕':1, '파이썬':133, '크롤링':5, '시각적':3}

wc = wordcloud.WordCloud(font_path='C:/Windows/Fonts/batang.ttc')
cloud = wc.generate_from_frequencies(keywords)

plt.figure()
plt.imshow(cloud)
plt.show()
```

```python
# 배경색상 및 크기 변경
import wordcloud
import matplotlib.pyplot as plt
%matplotlib inline
keywords = {'안녕':5, '파이썬':30, '크롤링':5, '시각적':3}

wc = wordcloud.WordCloud(font_path='C:/Windows/Fonts/batang.ttc', background_color='white', width=1000, height=800)
cloud = wc.generate_from_frequencies(keywords)

plt.figure()
plt.imshow(cloud)
plt.show()

## savefig() 활용, 이미지를 파일로 저장할 수도 있음
# figure.savefig('wordCloud.png')
```
